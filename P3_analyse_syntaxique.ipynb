```python
import spacy
from spacy import displacy

# Download the French model if it's not already installed
!python -m spacy download fr_core_news_sm

# Chargement du modèle linguistique de spaCy
nlp = spacy.load("fr_core_news_sm")

# Texte à analyser
text = "La technologie de l'intelligence artificielle évolue rapidement."

# Analyse du texte
doc = nlp(text)

#affichage des dépendances syntaxiques
print("analyses syntaxiques :")
for token in doc:
                 print(f"{token.text:<12} {token.dep_:<10} {token.head.text:<10} {token.pos_}")

#Visualisation de l'arbre de dépendance
displacy.render(doc, style="dep", jupyter=True)
```
