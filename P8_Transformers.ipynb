```python
import spacy
from spacy import displacy
import pandas as pd

# Load the correct spaCy model
nlp = spacy.load('en_core_web_sm') # Changed model name to 'en_core_web_sm'
# Tell pandas to handle potentially malformed strings by escaping or ignoring errors
csv_msg = pd.read_csv('IMDB Dataset.csv', header=None, on_bad_lines='skip', quoting=3) #Added on_bad_lines='skip' and quoting=3
#quoting=3 tells the parser to use the QUOTE_NONE strategy which disables quoting. This can be useful if your CSV file uses a quoting style that Pandas doesn't recognize or if it has inconsistent quoting.
#on_bad_lines='skip' is added to automatically skip lines that cause parsing errors, preventing the code from abruptly stopping.
for msg in csv_msg[0]:
    doc = nlp(msg)
    print("Analyse syntaxique")
    for token in doc:
        print(f"{token.text:<10}, {token.pos}, {token.dep_:<10}")

displacy.serve(doc, style='dep')
```
